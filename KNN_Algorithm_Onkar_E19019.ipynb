```python
import os 
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
X = iris[['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']]
y = iris[['Species']]
```

```python
def euclidean(a,b):
    dist = np.sqrt(sum((a-b)**2))
    return dist
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.5, stratify = y, random_state = 123)
```

### For Classification

```python
#k = 3
y_pred = []

def KNN(X_train, X_test, y_train, k):

    for j in X_test.index:
        dist = []
        index = []
        y_vals = []

        dist2 = []

        for i in X_train.index:
            d = euclidean(X_test.loc[j,:],X_train.loc[i,:])

            dist.append(d)
            index.append(i)
            y_vals.append(y_train.loc[i])

            dist2.append((d,i,y_train.loc[i,'Species']))

        dist2.sort()

        vals = []

        for kkk in dist2[0:k]:
            vals.append(kkk[2])

        #print(vals)

        m = vals
        l = {}
        for vals in m:
            if vals not in l:
                l[vals] = 1
            else:
                l[vals] += 1
        #print(l)

        list_keys = [k for k in l]
        #print(list_keys)
        y_pred.append((list_keys[0],j))
        
    return y_pred
```

```python
KNN(X_train, X_test, y_train, 1)
```
